## Setup

```python
import pandas as     pd
from   pandas import Series, DataFrame
import numpy  as     np
import matplotlib.pyplot as plt
import xlrd
#import seaborn as sns
```

```python
column_names = ['Year', 'Project', 'Entity', 'Country', 'Budget']
```

```python
#pd.set_option('display.height', 1000)
pd.set_option('display.max_colwidth',  600)
pd.set_option('display.max_rows',      500)
pd.set_option('display.max_columns',   900)
pd.set_option('display.width',        1500)
```

### Read data
File is 'EU-Data/export_2015_2017.csv' which is simply a slightly edited concatenation of files where annual budgets have been dumped, years 2015 to 2017

```python
df = pd.read_csv('5G-2015-2018.csv', delimiter=',' , names = column_names)

df.head(3)
```

## Data presentation

```python
df.groupby('Country')['Budget'].sum().nlargest(10)
```

```python
CountrySummary = df.groupby('Country')['Budget'].sum().nlargest(10)

plt.figure(figsize=(12, 10), dpi= 100);
plt.xlabel("Budget", fontsize=18)
plt.ylabel("Country", fontsize=18)

plt.grid()
plt.title("5G Budgets per country", fontsize=22)
plt.savefig('Plots/Accelerations.png')

CountrySummary.plot(kind='barh');
```
